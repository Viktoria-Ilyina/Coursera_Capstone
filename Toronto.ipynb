# The third task

```python
#Select Toronto data
toronto_data = df[df['Borough'].str.contains('Toronto')].reset_index(drop=True)
toronto_data
```

```python
#Install and import additional required libraries
!pip install folium
import folium
from sklearn.cluster import KMeans
import matplotlib.cm as cm
import matplotlib.colors as colors
```

```python
#Visualize all the Neighbourhoods of the above dataframe using Folium

map_toronto = folium.Map(location=[43.676357,-79.293031],zoom_start=11)

for lat,lng,borough,neighbourhood in zip(toronto_data['Latitude'],toronto_data['Longitude'],toronto_data['Borough'],toronto_data['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=4,
    popup=label,
    color='yellow',
    fill=True,
    fill_color='#6d85c9',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
map_toronto

```

```python
#Using KMeans clustering for the clsutering of the neighbourhoods
k=5
toronto_clustering = toronto_data.drop(['Postal Code','Borough','Neighbourhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_
toronto_data.insert(0, 'Cluster Labels', kmeans.labels_)
toronto_data
```

```python
# Create map
map_clusters = folium.Map(location=[43.676357,-79.293031],zoom_start=11)

# Set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighbourhood, cluster in zip(toronto_data['Latitude'], toronto_data['Longitude'], toronto_data['Neighbourhood'], toronto_data['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
